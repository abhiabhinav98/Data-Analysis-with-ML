```python
knn_data = data[["Gender","Occupation","Purchase"]]
x = data[["Occupation","Purchase"]]
y = data.Gender
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
xx = x_test[3:4]
model.predict(xx)
```

```python
model.score(x_train, y_train)
```

```python
n = np.arange(1,30)
test_accuracy = []
train_accuracy = []
for i,k in enumerate(n):
    knn_clf = KNeighborsClassifier(n_neighbors=k)
    knn_clf.fit(x_train,y_train)
    train_accuracy.append(knn_clf.score(x_train,y_train))
    test_accuracy.append(knn_clf.score(x_test,y_test))
    print("Step:", k)
    
plt.figure(figsize=(10,15))    
plt.plot(n, train_accuracy, label = 'Testing Accuracy')
plt.plot(n, test_accuracy, label = 'Training Accuracy')
plt.legend()

    
    
```
