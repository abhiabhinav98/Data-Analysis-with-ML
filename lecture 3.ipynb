```python
import matplotlib.pyplot as plt
import numpy as np
import random
fig = plt.figure(figsize=(10,10))

ax1 = plt.subplot2grid((6,2),(0,0), rowspan =1, colspan=2)
ax2 = plt.subplot2grid((6,2),(1,1), rowspan =4, colspan=1)
ax3 = plt.subplot2grid((6,2),(5,0), rowspan =1, colspan=2)

def create_plots():
    xs = []
    yx = []
    for i in range(10):
        x = i
        y = random.randrange(10)
        xs.append(i)
        yx.append(y)
    return xs, yx

x, y = create_plots()
ax1.plot(x,y)
x, y = create_plots()
ax2.plot(x,y)
x, y = create_plots()
ax3.plot(x,y)

plt.show()
```
