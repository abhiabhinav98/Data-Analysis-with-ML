```python
x = [1,2,3,4,5] #Temp
y = [34,56,7,9,90] #Temp
plt.scatter(x,y)
```

```python
x = [1,2,3,4,5] #Temp
y = [34,56,7,9,90] #Temp
plt.scatter(x,y)
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("mtlib")
plt.show()
```

```python
x = [1,2,3,4,5] #Temp
y = [34,56,7,9,90] #Temp
colors = ["red","blue","green","yellow","red"]
plt.scatter(x,y, c=colors)
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("mtlib")
plt.show()
```

```python
x = df["latitude"]
y = df["longitude"]
plt.scatter(x,y)
plt.show()
```

```python
df["median_house_value"].hist(bins=5)
plt.show()
```

```python
df.hist(bins=5)
plt.show()
```

```python
df.hist(bins=5, figsize=(10,8))
plt.show()
```

```python
df.plot(x = "longitude", y = "latitude")
plt.show()
```

```python
df.plot(kind = "scatter", x = "longitude", y = "latitude")
plt.show()
```

```python
df.plot(kind = "scatter", x = "longitude", y = "latitude", c="median_house_value")
plt.show()
```

```python
df.plot(kind = "scatter", x = "longitude", y = "latitude", c="median_house_value", cmap = plt.get_cmap("jet"))
plt.show()
```

```python
df.plot(kind = "scatter", x = "longitude", y = "latitude",s=df["population"], c="median_house_value", cmap = plt.get_cmap("jet"))
plt.show()
```

```python
df.plot(kind = "scatter", x = "longitude", y = "latitude",s=df["population"]/100, c="median_house_value", cmap = plt.get_cmap("jet"))
plt.show()
```

```python
df.plot(kind = "scatter", x = "longitude", y = "latitude",s=df["population"]/100, c="median_house_value", cmap = plt.get_cmap("jet"), figsize=(15,10))
plt.show()
```
